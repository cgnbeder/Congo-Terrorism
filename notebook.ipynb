```python
import pandas as pd
terrorism = pd.read_excel("terrordata.xlsx")
terrorism = terrorism[["iyear", "country", "country_txt", "region", "region_txt", "city", "latitude", "longitude", "specificity", "vicinity", "alternative", "multiple", "success", "attacktype1", "attacktype1_txt", "targtype1", "targtype1_txt", "targsubtype1", "targsubtype1_txt", "target1", "natlty1", "natlty1_txt", "weaptype1", "weaptype1_txt", "weapsubtype1", "weapsubtype1_txt", "weapdetail", "nkill", "nkillus", "nwound", "nwoundus"]]
```

```python
# Terror attacks by year/country
terrorism_year_country_nkill_nwound = terrorism.groupby(["iyear", "country_txt"])[["nkill", "nwound"]].sum().reset_index()
terrorism_year_country_nkill_nwound.loc[terrorism_year_country_nkill_nwound["country_txt"] == "Democratic Republic of the Congo"]
terrorism_year_country_nkill_nwound
```

```python
# Export terrorism types by country and year, filtered to only include data since 2010. Then take a sum.
terrorism_year_country_attacktype = terrorism.groupby(["iyear", "country_txt"])["attacktype1_txt"].value_counts().reset_index(name='count')
terrorism_year_country_attacktype = terrorism_year_country_attacktype[terrorism_year_country_attacktype["iyear"] > 2010]

# Also filter for D.R. Congo's border neighbors
border_countries = ["Angola", "Burundi", "Cameroon", "Central African Republic", "Democratic Republic of the Congo", "Kenya", "Rwanda", "South Sudan", "Tanzania", "Zambia"]
terrorism_year_country_attacktype_border_countries = terrorism_year_country_attacktype[terrorism_year_country_attacktype["country_txt"].isin(border_countries)]
terrorism_year_country_attacktype_border_countries_sum = terrorism_year_country_attacktype_border_countries.groupby(["country_txt", "attacktype1_txt"]).sum().reset_index()
terrorism_year_country_attacktype_border_countries_sum = terrorism_year_country_attacktype_border_countries_sum[["country_txt", "attacktype1_txt", "count"]]
terrorism_year_country_attacktype_border_countries_sum
```

```python
# African countries GDP comp.
africa_gdp = pd.read_excel("africa_gdp.xlsx")
first_three_column = africa_gdp.iloc[:, :3]
last_columns = africa_gdp.iloc[:, 14:]
africa_gdp_after_1970 = pd.concat([first_three_column, last_columns], axis=1)
africa_gdp_after_1970
```

```python
# Imported country codes to select only African countries
africa_country_codes = [
    "DZA", "AGO", "BEN", "BWA", "BFA", "BDI", "CPV", "CMR", "CAF", "TCD",
    "COM", "COG", "COD", "DJI", "EGY", "GNQ", "ERI", "SWZ", "ETH", "GAB",
    "GMB", "GHA", "GIN", "GNB", "CIV", "KEN", "LSO", "LBR", "LBY", "MDG",
    "MWI", "MLI", "MRT", "MUS", "MAR", "MOZ", "NAM", "NER", "NGA", "RWA",
    "STP", "SEN", "SYC", "SLE", "SOM", "ZAF", "SSD", "SDN", "TZA", "TGO",
    "TUN", "UGA", "ZMB", "ZWE"]

africa_gdp_after_1970_selected = africa_gdp_after_1970[africa_gdp_after_1970["Country Code"].isin(africa_country_codes)]
africa_gdp_after_1970_selected
```

```python
# Needed to melt data to see years by row
africa_gdp_after_1970_selected_melted = africa_gdp_after_1970_selected.melt(id_vars=["Country Name", "Country Code", "Indicator Name"], 
                    var_name="Year", 
                    value_name="GDP (current US$)")
africa_gdp_after_1970_selected_melted.dropna(axis=0, inplace=True)
africa_gdp_after_1970_selected_melted[africa_gdp_after_1970_selected_melted["Country Code"] == "COD"]
```
